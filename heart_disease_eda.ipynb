```python
df = df[(df['Cholesterol'] != 0) & (df['RestingBP'] != 0)]
df.shape
```

```python
sns.countplot(data=df, x='HeartDisease')
plt.title('Distribution of Heart Disease')
plt.savefig(f"{save_path}\\heart_disease_distribution.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
sns.countplot(data=df, x='ChestPainType', hue='HeartDisease')
plt.title('Chest Pain Type vs Heart Disease')
plt.savefig(f"{save_path}\\chest_pain_vs_heart_disease.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
sns.countplot(data=df, x='ExerciseAngina', hue='HeartDisease')
plt.title('Exercise-Induced Angina vs Heart Disease')
plt.savefig(f"{save_path}\\exercise_angina_vs_heart_disease.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
sns.boxplot(data=df, x='HeartDisease', y='MaxHR')
plt.title('Max Heart Rate vs Heart Disease')
plt.savefig(f"{save_path}\\maxhr_vs_heart_disease.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
sns.boxplot(data=df, x='HeartDisease', y='Oldpeak')
plt.title('Oldpeak vs Heart Disease')
plt.savefig(f"{save_path}\\oldpeak_vs_heart_disease.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
sns.countplot(data=df, x='ST_Slope', hue='HeartDisease')
plt.title('ST Slope vs Heart Disease')
plt.savefig(f"{save_path}\\st_slope_vs_heart_disease.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.savefig(f"{save_path}\\correlation_matrix.png", dpi=300, bbox_inches='tight')
plt.show()
```
